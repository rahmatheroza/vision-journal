In this notebook, you'll learn how to manipulate images using OpenCV.

Let's build your image processing toolbox!


```python
import cv2
import numpy as np
import matplotlib.pyplot as plt

%matplotlib inline

# Load image
img = cv2.imread('/content/teddy.jpeg')
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
img_rgb.shape
```

```python
# Resize to 500x500
resized = cv2.resize(img_rgb, (500, 500))

# Crop Teddy with the same size 500x500
cropped = img_rgb[500:1000, 200:700]

# Show both
plt.figure(figsize=(8, 4))
plt.subplot(1, 2, 1)
plt.imshow(resized)
plt.title('Resized (500x500)')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(cropped)
plt.title('Cropped (500x500)')
plt.axis('off')
plt.show()
```

```python
# Rotate 90Â° clockwise
rotated = cv2.rotate(img_rgb, cv2.ROTATE_90_CLOCKWISE)

# Flip horizontally
flipped = cv2.flip(img_rgb, 1)

plt.figure(figsize=(10, 4))
plt.subplot(1, 2, 1)
plt.imshow(rotated)
plt.title('Rotated 90Â°')
plt.axis('off')

plt.subplot(1, 2, 2)
plt.imshow(flipped)
plt.title('Flipped Horizontally')
plt.axis('off')
plt.show()
```

```python
# convert to grayscale
gray = cv2.cvtColor(img_rgb, cv2.COLOR_RGB2GRAY)

plt.imshow(gray, cmap='gray')
plt.title('Grayscale')
plt.axis('off')
plt.show()

print("Gray shape:", gray.shape)  # 2D array
```

```python
# Make a copy to draw on
img_draw = img_rgb.copy()

# Draw a rectangle
cv2.rectangle(img_draw, (200, 500), (700, 1000), (255, 0, 0), thickness=2)

# Draw a circle
cv2.circle(img_draw, (180, 360), 40, (0, 255, 0), thickness=3)

# Add text
cv2.putText(img_draw, 'Hello!', (50, 460), cv2.FONT_HERSHEY_SIMPLEX, 2, (0, 0, 255), 2)

plt.imshow(img_draw)
plt.title('Drawing Shapes & Text')
plt.axis('off')
plt.show()
```

```python
# Apply Gaussian blur
blurred = cv2.GaussianBlur(img_rgb, (11, 11), 0)

plt.imshow(blurred)
plt.title('Blurred Image (Gaussian)')
plt.axis('off')
plt.show()
```
