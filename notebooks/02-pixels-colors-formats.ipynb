```python
# üõ†Ô∏è Import Libraries
import cv2
import numpy as np
import matplotlib.pyplot as plt

# Enable inline plotting for notebooks
%matplotlib inline
```

```python
# Load an image (You can upload your own using Colab's file uploader)
img = cv2.imread('/content/playground.jpeg')  # Make sure the image is in the same directory
print("Image shape:", img.shape)

# Convert BGR (OpenCV default) to RGB for correct display
img_rgb = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# Show the image
plt.imshow(img_rgb)
plt.title('Original Image')
plt.axis('off')
plt.show()
```

```python
# Check pixel value at a specific location (e.g., top-left corner)
pixel = img[0, 0]
print("BGR pixel value at (0, 0):", pixel)

# Each pixel has 3 values (Blue, Green, Red)
b, g, r = pixel
print(f"Blue: {b}, Green: {g}, Red: {r}")
```

```python
# Split the RGB channels
r, g, b = cv2.split(img_rgb)

# Plot each channel separately
plt.figure(figsize=(12,4))

plt.subplot(1, 3, 1)
plt.imshow(r, cmap='Reds')
plt.title('Red Channel')
plt.axis('off')

plt.subplot(1, 3, 2)
plt.imshow(g, cmap='Greens')
plt.title('Green Channel')
plt.axis('off')

plt.subplot(1, 3, 3)
plt.imshow(b, cmap='Blues')
plt.title('Blue Channel')
plt.axis('off')

plt.tight_layout()
plt.show()
```

```python
# Save image in different formats
cv2.imwrite('playground_compressed.jpg', img, [int(cv2.IMWRITE_JPEG_QUALITY), 50])
cv2.imwrite('playground_lossless.png', img)
cv2.imwrite('playground_raw.bmp', img)

# Reload them
jpg = cv2.imread('playground_compressed.jpg')
png = cv2.imread('playground_lossless.png')
bmp = cv2.imread('playground_raw.bmp')

# Convert all to RGB
jpg = cv2.cvtColor(jpg, cv2.COLOR_BGR2RGB)
png = cv2.cvtColor(png, cv2.COLOR_BGR2RGB)
bmp = cv2.cvtColor(bmp, cv2.COLOR_BGR2RGB)

# Display side by side
titles = ['JPEG (compressed)', 'PNG (lossless)', 'BMP (raw)']
images = [jpg, png, bmp]

plt.figure(figsize=(15,5))
for i in range(3):
    plt.subplot(1, 3, i+1)
    plt.imshow(images[i])
    plt.title(titles[i])
    plt.axis('off')
plt.tight_layout()
plt.show()
```
