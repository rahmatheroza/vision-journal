### ðŸ”„ Simple Image Processing Pipeline

This is a **basic pipeline** where the algorithm follows this flow:

Read Image âž¡ï¸ Preprocess (convert RGB image to Grayscale) âž¡ï¸ Output

```python
# Load and display an image using OpenCV and Google Colab's display tools
import cv2
from google.colab.patches import cv2_imshow  # Colab-specific
from IPython.display import display
from PIL import Image
import numpy as np

# Read image
image = cv2.imread('/content/lion-walk.jpg')

# Convert to grayscale
gray = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Display using Colab's cv2_imshow
print("Original Image:")
cv2_imshow(image)

print("Grayscale Image:")
cv2_imshow(gray)
```
